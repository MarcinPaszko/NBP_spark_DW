```python
import yfinance as yf
from pathlib import Path
import pandas as pd
import time

# Listy tickerÃ³w
groups = {
    "USA": ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'NVDA', 'META', 'TSLA', 'BRK-B', 'UNH', 'V', 'JNJ', 'WMT', 'JPM', 'MA', 'PG', 'AVGO', 'HD', 'CVX', 'ORCL', 'ABBV', 'KO', 'PEP', 'COST', 'BAC', 'ADBE'],
    "INDEX": ['^GSPC', '^IXIC', '^DJI', '^RUT', '^VIX', '^FTSE', '^GDAXI', '^FCHI', '^N225', '^HSI', '^STOXX50E', 'QQQ', 'SPY', 'IWM', 'EEM', 'VWO', 'VEA', 'VNQ', 'GLD', 'SLV', 'DBC', 'TLT', 'HYG', 'VTI', 'VXUS'],
    "GPW": ['CDR.WA', 'PKO.WA', 'PKN.WA', 'KGH.WA', 'PZU.WA', 'PEA.WA', 'ALE.WA', 'LPP.WA', 'DNP.WA', 'PGE.WA', 'OPL.WA', 'SPL.WA', 'JSW.WA', 'ACP.WA', 'KRU.WA', 'MBK.WA', 'CPS.WA', 'TPE.WA', 'KTY.WA', 'ATT.WA', 'ASB.WA', 'BDX.WA', 'GPW.WA', '11B.WA', 'CCC.WA'],
    "CRYPTO": ['BTC-USD', 'ETH-USD', 'SOL-USD', 'BNB-USD', 'XRP-USD', 'ADA-USD', 'DOGE-USD', 'AVAX-USD', 'DOT-USD', 'TRX-USD', 'LINK-USD', 'MATIC-USD', 'SHIB-USD', 'LTC-USD', 'DAI-USD', 'BCH-USD', 'ATOM-USD', 'UNI-USD', 'XLM-USD', 'XMR-USD', 'ETC-USD', 'FIL-USD', 'HBAR-USD', 'APT-USD', 'NEAR-USD'],
    "COMMODITIES": ['GC=F', 'SI=F', 'CL=F', 'NG=F', 'HG=F', 'ZC=F', 'ZW=F', 'KC=F', 'CC=F', 'CT=F', 'OJ=F', 'PL=F', 'PA=F', 'SB=F', 'SOYB', 'WEAT', 'CORN', 'WOOD', 'LIT', 'REMX', 'TAN', 'FAN', 'PICK', 'URA', 'JO']
}

ALL_TICKERS = [item for sublist in groups.values() for item in sublist]
STOCKS_PATH = Path("/home/jovyan/work/data/bronze/historical_data_stocks")
STOCKS_PATH.mkdir(parents=True, exist_ok=True)

print(f"ðŸš€ Rozpoczynam pobieranie {len(ALL_TICKERS)} tickerÃ³w...")

for ticker in ALL_TICKERS:
    # Nazwa pliku - zamieniamy znaki specjalne, by Windows nie protestowaÅ‚
    safe_name = ticker.replace("^", "IDX_").replace("=", "FUT_").replace("-", "_")
    file_path = STOCKS_PATH / f"{safe_name}.csv"
    
    if not file_path.exists():
        try:
            # Pobieramy dane
            df = yf.download(ticker, start="2020-01-01", end="2025-12-31", progress=False)
            if not df.empty:
                # Dodajemy kolumnÄ™ z tickerem, bo Yahoo jej nie eksportuje do CSV domyÅ›lnie
                df['Ticker'] = ticker
                df.to_csv(file_path)
                print(f"âœ… {ticker} pobrany.")
                time.sleep(0.2) # MaÅ‚a pauza, by nie dostaÄ‡ bana od Yahoo
        except Exception as e:
            print(f"âŒ BÅ‚Ä…d {ticker}: {e}")

print("\nðŸ“¦ Wszystkie pliki CSV sÄ… w folderze bronze!")
```

Input into DB

```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pathlib import Path

# 1. Inicjalizacja sesji Spark
spark = SparkSession.builder \
    .appName("Bulk_Stocks_Ingest_Fixed") \
    .config("spark.jars.packages", "org.postgresql:postgresql:42.6.0") \
    .getOrCreate()

STOCKS_PATH = "/home/jovyan/work/data/bronze/historical_data_stocks"

print("ðŸ“‚ WczytujÄ™ pliki CSV...")

# 2. Odczyt danych z CSV
raw_stocks_df = spark.read.option("header", "true") \
    .option("inferSchema", "true") \
    .csv(f"{STOCKS_PATH}/*.csv")

# PodglÄ…d - sprawdÅºmy czy kolumna 'Price' to faktycznie daty
print("ðŸ‘€ PodglÄ…d surowych danych:")
raw_stocks_df.select("Ticker", "Price").show(5)

# 3. Mapowanie, konwersja typÃ³w i czyszczenie
# UÅ¼ywamy F.to_date na kolumnie 'Price', bo Spark jÄ… tak nazwaÅ‚ w Twoich plikach
final_stocks = raw_stocks_df.select(
    F.col("Ticker").alias("ticker"),
    F.to_date(F.col("Price")).alias("trade_date"), 
    F.col("Open").cast("decimal(12,2)").alias("open_price"),
    F.col("High").cast("decimal(12,2)").alias("high_price"),
    F.col("Low").cast("decimal(12,2)").alias("low_price"),
    F.col("Close").cast("decimal(12,2)").alias("close_price"),
    F.col("Volume").cast("bigint").alias("volume")
).filter(
    (F.col("close_price") > 0) & 
    (F.col("ticker").isNotNull()) &
    (F.col("trade_date").isNotNull())
)

# 4. Zapis do Postgresa
print(f"ðŸ’¾ ZapisujÄ™ {final_stocks.count()} rekordÃ³w do tabeli f_stock_prices...")

final_stocks.write \
    .format("jdbc") \
    .option("url", "jdbc:postgresql://postgres_dw:5432/currency_db") \
    .option("dbtable", "f_stock_prices") \
    .option("user", "admin") \
    .option("password", "password123") \
    .option("driver", "org.postgresql.Driver") \
    .mode("overwrite") \
    .save()

print("\nâœ¨ FINALNY SUKCES! Dane gieÅ‚dowe sÄ… w bazie.")

# KrÃ³tkie podsumowanie na koniec
final_stocks.groupBy("ticker").count().orderBy(F.desc("count")).show(10)
```
